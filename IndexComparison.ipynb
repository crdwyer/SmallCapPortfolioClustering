# sp500 vs russell2000 index performance
Data is obtained from yahoo.finance (manual csv download)



https://finance.yahoo.com/quote/%5ERUT/history?p=%5ERUT



https://finance.yahoo.com/quote/%5EGSPC/history/

```python
rut = pd.read_csv(r"RUT.csv")
rut_month = pd.read_csv(r"RUT_month.csv")
rut['Date']= pd.to_datetime(rut['Date'])
rut_month['Date']= pd.to_datetime(rut_month['Date'])
```

```python
rut['returns'] = rut['Adj Close'].pct_change()
rut_month['returns'] = rut_month['Adj Close'].pct_change()
```

```python
sp_month = pd.read_csv(r"GSPC.csv")
sp = pd.read_csv(r"GSPC_daily.csv")
sp['Date']= pd.to_datetime(sp['Date']) 
sp_month['Date']= pd.to_datetime(sp_month['Date']) 
sp_month.sort_values(by=['Date'], inplace=True, ascending=True)
sp.sort_values(by=['Date'], inplace=True, ascending=True)
```

```python
sp_month['Returns'] = sp_month['Adj Close'].pct_change()
sp['Returns'] = sp['Adj Close'].pct_change()
```

```python
import plotly.express as px
fig = px.line(x= sp_month['Date'] ,y= [sp_month['Returns'],rut_month['returns']] ,title="Small Cap vs Large Cap Index Returns")
fig.update_yaxes(title_text ="Returns")
fig.update_xaxes(title_text = "Date")
newnames = {'wide_variable_0':'SP 500', 'wide_variable_1': 'Russell 2000'}

fig.for_each_trace(lambda t: t.update(name = newnames[t.name]))
fig.update_layout(legend_title_text="Index")
fig.show()
```
